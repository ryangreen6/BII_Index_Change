```python
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import geopandas as gpd
import rioxarray as rioxr
from shapely import box

import pystac_client
import planetary_computer

from IPython.display import Image 
```

```python
az = gpd.read_file("data/tl_2024_04_cousub.shp")

phoenix = az[az["NAME"] == "Phoenix"]

phoenix.boundary.plot()
```

```python
phoenix_bbox = list([-112.826843, 32.974108, -111.184387, 33.863574])
```

```python
catalog = pystac_client.Client.open(
    "https://planetarycomputer.microsoft.com/api/stac/v1",
    modifier=planetary_computer.sign_inplace,
)

search = catalog.search(collections = ['io-biodiversity'],
                       bbox = phoenix_bbox)
search
```

```python
items = search.item_collection()
print(f"Returned {len(items)} Items")
items
```

```python
item2017 = items[3]

Image(url = item2017.assets['rendered_preview'].href, width = 600)
```

```python
BII2017 = rioxr.open_rasterio(item2017.assets['data'].href)
BII2017
```

```python
item2020 = items[0]

Image(url = item2020.assets['rendered_preview'].href, width = 600)
```

```python
BII2020 = rioxr.open_rasterio(item2020.assets['data'].href)
BII2020
```

```python
BII2017 = BII2017.squeeze().drop_vars('band')
```

```python

phoenix_match = phoenix.to_crs(BII2017.rio.crs)
assert phoenix_match.crs == BII2017.rio.crs

BII2017 = BII2017.rio.write_nodata(0)

BII2017 = (BII2017.rio.clip_box(*phoenix_match.total_bounds)
              .rio.clip(phoenix_match.geometry))

BII2017.plot()
```

```python
BII2020 = BII2020.squeeze().drop_vars('band')
```

```python
BII2020 = BII2020.rio.write_nodata(0)

BII2020 = (BII2020.rio.clip_box(*phoenix_match.total_bounds)
              .rio.clip(phoenix_match.geometry))

BII2020.plot()
```

```python
BII2020_75 = BII2020>0.75

BII2020_75.plot()
```

```python
BII2017_75 = BII2017>0.75
BII2017_75.plot()
```

```python
BII2017_75 = BII2017_75.fillna(0)
BII2020_75 = BII2020_75.fillna(0)

diff = BII2017_75 - BII2020_75

```

```python

diff.plot()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

BII2020.plot(ax=ax, cmap='viridis', alpha=0.8)

diff.plot(ax=ax, cmap = "Reds", levels=10, linewidths=0.5)

ax.set_title('BII2017 with diff overlay')
plt.show()
```
