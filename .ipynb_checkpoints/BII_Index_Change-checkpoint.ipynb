```python
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

import geopandas as gpd
import rioxarray as rioxr
import contextily as cx
from shapely import box

import pystac_client
import planetary_computer

from IPython.display import Image 
```

```python
az = gpd.read_file("data/tl_2024_04_cousub.shp")

phoenix = az[az["NAME"] == "Phoenix"]

phoenix.boundary.plot()
```

```python
phoenix_bbox = list([-112.826843, 32.974108, -111.184387, 33.863574])
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

phoenix.boundary.plot(ax=ax, color = 'black')

cx.add_basemap(ax=ax, crs = phoenix.crs)

ax.set_title('Biodiversity Intactness Index (BII)\nPhoenix Subdivision')
ax.axis('off')
plt.show()
```

```python
catalog = pystac_client.Client.open(
    "https://planetarycomputer.microsoft.com/api/stac/v1",
    modifier=planetary_computer.sign_inplace,
)

search = catalog.search(collections = ['io-biodiversity'],
                       bbox = phoenix_bbox)
search
```

```python
items = search.item_collection()
print(f"Returned {len(items)} Items")
items
```

```python
item2017 = items[3]

BII2017 = rioxr.open_rasterio(item2017.assets['data'].href)
BII2017
```

```python
item2020 = items[0]

BII2020 = rioxr.open_rasterio(item2020.assets['data'].href)
BII2020
```

```python
BII2017 = BII2017.squeeze().drop_vars('band')

BII2017 = (BII2017.rio.clip_box(*phoenix.total_bounds)
              .rio.clip(phoenix.geometry))
```

```python
BII2020 = BII2020.squeeze().drop_vars('band')

BII2020 = (BII2020.rio.clip_box(*phoenix.total_bounds)
              .rio.clip(phoenix.geometry))
```

```python
BII2020_75 = BII2020 >= 0.75
BII2017_75 = BII2017 >= 0.75
```

```python
BII2017_75 = BII2017_75.fillna(0)
BII2020_75 = BII2020_75.fillna(0)

diff = BII2017_75.astype('int') - BII2020_75.astype('int')

diff = diff.where(diff == 1)
```

```python
diff.plot()
```

```python
total_pixels = BII2020 > 0

values, counts = np.unique(total_pixels, return_counts = True)
assert len(values) == len(counts)

x = {'values': values, 'counts': counts}
pixels = pd.DataFrame(data = x)
```

```python
values, counts = np.unique(BII2017_75, return_counts = True)
assert len(values) == len(counts)

x = {'values': values, 'counts': counts}
BII2017_75pix = pd.DataFrame(data = x)
pct_2017 = (BII2017_75pix.iloc[1,1] / pixels.iloc[1,1]) * 100
```

```python
values, counts = np.unique(BII2020_75, return_counts = True)
assert len(values) == len(counts)

x = {'values': values, 'counts': counts}
BII2020_75pix = pd.DataFrame(data = x)
pct_2020 = (BII2020_75pix.iloc[1,1] / pixels.iloc[1,1]) * 100
```

```python
print(pct_2017)
print(pct_2020)
```

```python
fig, ax = plt.subplots(figsize=(8, 5))


BII2020.plot(ax=ax, 
             cmap='Greens', 
             alpha=0.8,
             cbar_kwargs={'orientation': 'horizontal',
                         'label': 'BII for 2020'})

phoenix.boundary.plot(ax=ax, 
                            color = 'black')

diff.plot(ax=ax, 
          cmap = 'gist_heat', 
          linewidths=0.5, add_colorbar = False)

red_patch = mlines.Line2D([], [], color='red', label='Area with BII ≥ 0.75 lost from 2017 to 2020', linewidth=4)

ax.legend(handles=[red_patch], bbox_to_anchor = (0, 0, 0.9, 0), fontsize=8)

ax.set_title('Biodiversity Intactness Index (BII)\nPhoenix Subdivision')
ax.axis('off')
plt.show()
```
